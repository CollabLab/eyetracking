```python
import pickle
```

```python
import time
```

```python
playback_file = open('playback', 'rb') # replace for relevant  file name
data2 = pickle.load(playback_file)
```

```python
def play():
    initial_time = min(data2['eyetrack_data'][0][5], data2['moving_object_data'][0][4])
    end_time = max(data2['eyetrack_data'][-1][5], data2['moving_object_data'][-1][4])
    total_time = end_time - initial_time

    start_time = int(round(time.time() * 1000))
    while True:
        playback_now = (int(round(time.time() * 1000))) - start_time
        #print("playback : %s" % playback_now)
        if playback_now >= total_time:
            break
        mo_coords = data2['moving_object_list'][playback_now]

        if mo_coords is not None: # do something if coordinates are there
            print("playback : %s, %s" % mo_coords)
```

```python
play()
```
